0. Импорт необходимых библиотек и скачивание всех зависимостей

```python
from string import punctuation

import pandas as pd
import pymorphy2
import nltk

nltk.download('punkt')
nltk.download('stopwords')
```

1. Зададим несколько предложений, содержащих знаки препинания и заглавные буквы

```python
sentences = [
    'Привет, Я ChatGPT, крупная языковая модель, обученная на архитектуре GPT-3.5 от OpenAI.',
    'Я использую искусственный интеллект для того, чтобы отвечать на вопросы и помогать людям по всему миру.',
    'Моя обученность позволяет мне понимать язык, генерировать текст и решать различные задачи.'
]
```

2.1. Напишем функцию для предобработки переданной на вход строки

```python
def preprocess_string(input_string: str) -> list[str]:
    tokens = nltk.tokenize.word_tokenize(input_string)  # токенизируем строку
    token_no_punct = list(filter(lambda t: t not in punctuation, tokens))  # отсекаем пунктуацию
    low_tokens = list(map(lambda t: t.lower(), token_no_punct))  # приводим к нижнему регистру

    stop_words = nltk.corpus.stopwords.words('russian')  # из модуля nltk получаем стоп-слова
    tokens_no_stop = list(filter(lambda t: t not in stop_words, low_tokens))  # отсекаем стоп-слова

    morph = pymorphy2.MorphAnalyzer()  # инициализируем морфологический анализатор
    lemms = list(map(lambda t: morph.parse(t)[0].normal_form, tokens_no_stop))  # проводим лемматизацию

    return lemms  # возвращаем полученный список токенов
```

2. Проведём предобработку предложений из п.1 и выведем результат

```python
preprocessed_sentences = list(map(lambda s: preprocess_string(s), sentences))  # применяем всё вплоть до лемматизации

stemmer = nltk.stem.snowball.RussianStemmer()  # инициализируем стеммер
stems = [[stemmer.stem(l) for l in s] for s in preprocessed_sentences]  # дополнительно проводим стемминг

stems
```

4-5. Выполненим предобработку всех текстов и запишем получившееся в отдельную колонку

```python
db['Transcription tokens'] = db['Transcription'].apply(lambda tr: ' '.join(preprocess_string(tr)))
db
```
